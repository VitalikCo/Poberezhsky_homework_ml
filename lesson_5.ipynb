# Градиентный спуск

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Функция одной переменной

```python
# Функция одной переменной
def f_x(x):
    return 3 * x**2
```

```python
# зададим точки по оси x 
x = np.linspace(-5, 5, 1000)

# рассчитаем y
y = f_x(x)

plt.plot(x, y)

plt.title('y = 3 * x^2')

plt.show()
```

```python
# производная функции y = 3 * x^2
def df_x(x):
    return 6 * x
```

```python
x0 = 5 # начальное значение по оси x, откуда начинаем градиентный спуск
la = 0.1 # шаг
```

```python
x1 = x0 - la * df_x(x0) # следующее значение по оси x
```

```python
res_x = []
res_fx = []
```

```python
while np.sqrt((x0 - x1)**2) > 0.00000001:
    res_x.append(x0)
    res_fx.append(f_x(x0))
    x0 = x1
    x1 = x0 - la * df_x(x0)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y)
plt.plot(res_x, res_fx, 'o-')
plt.legend(['y = 3 * x^2', 'Градиентный спуск'])
plt.show()
```

## Функция двух переменных

```python
# Функция двух переменных
def f_xy(x, y):
    return 3 * x ** 2 + 5 * y ** 2

# зададим точки по оси x 
x = np.linspace(-5, 5, 1000)

# Градиент (вектор частных производных)
def df_xy(x, y):
    return np.array([6 * x, 10 * y])

x0 = np.array([5, 10]) # координаты начального значение, откуда начинаем градиентный спуск
la = 0.1 # шаг

x1 = x0 - la * df_xy(x0[0], x0[1]) # координаты следующего значения

res_x = []
res_fx = []

while np.sqrt((x1 - x0)[0]**2 + (x1 - x0)[1]**2) > 0.00000001:
    res_x.append(x0)
    res_fx.append(f_xy(x0[0], x0[1]))
    x0 = x1
    x1 = x0 - la * df_xy(x0[0], x0[1])
```
